```python
# here we need to drop few variables, becoz while we encode these 'object' variables,there will be large no of columns
# for example 'Name' has more than 1800 values so when we encode, there will be 1800 columns..
# so this will affect the effeciency of model...
```

```python
# get_dummy encoding

a1=pd.get_dummies(a[['Location','Fuel_Type','Transmission','Owner_Type']],drop_first=True)
a1
```

```python
# now we have two dataframe a1 and a, we need to have dataframe which includes both the elements
# so we need to join or to be specifc 'concatenation'

a2=pd.concat([a,a1],axis=1) # axis=1 means 'column wise'
a2
```

```python
a3=a2.drop(['Unnamed: 0','Name','Location','Fuel_Type','Transmission','Owner_Type','New_Price'],axis=1)
a3
```

```python
a3.dtypes
```

```python
# there are units inside 'Mileage','Engine,'Power', we need to remove those units so replace it with ''

a3['Mileage']=a3['Mileage'].str.replace('km/kg','')
a3['Mileage']=a3['Mileage'].str.replace('kmpl','')
a3['Engine']=a3['Engine'].str.replace('CC','')
a3['Power']=a3['Power'].str.replace('bhp','')

# when we check after converting these units with '' still the dtypes of 'Mileage','Engine,'Power' are 'objects'
# it is shown 'objects' because it has 'null' values, so we are going to replace those null values with '0'

a3['Mileage']=a3['Mileage'].str.replace('null','0')
a3['Engine']=a3['Engine'].str.replace('null','0')
a3['Power']=a3['Power'].str.replace('null','0')
```

```python
a3
```

```python
a3.dtypes
```

```python
# but still is shown 'object', the model can't operate with 'objects' it should have only 'integers'
# so now we are going to change the 'datatype conversions'

a3['Mileage']=a3['Mileage'].astype(float)
a3['Power']=a3['Power'].astype(float)
a3['Engine']=a3['Engine'].astype(float)
```

```python
a3.dtypes
```

```python
# we filled the 'null' values with '0', 0 is a wrong information for any vehicles, there are no vehicles with 0 mileage

a3.loc[a3.Engine==0,'Engine']=np.NaN  # wherever 'Engine' has '0' value, it is replaced with NaN(Not a number)
a3.loc[a3.Power==0,'Power']=np.NaN    # wherever 'Power' has '0' value, it is replaced with NaN(Not a number
a3.loc[a3.Mileage==0,'Mileage']=np.NaN  # wherever 'Mileage' has '0' value, it is replaced with NaN(Not a number
```

```python
a3.isna().sum()
```

```python
# Handling Missing values

# 'seats' are filled with  'mode'
# the rest missing data is filled with 'mean'


a3[ "Engine"]=a3["Engine"].fillna(a3["Engine"].mean())
a3["Mileage"]=a3["Mileage"].fillna(a3["Mileage"].mean())
a3['Power']=a3['Power'].fillna(a3['Power'].mean())
a3['Seats']=a3['Seats'].fillna(a3['Seats'].mode()[0])
```

```python
a3.isna().sum() # hence no missing value
```

```python
# seperating x and y

x_train=a3.drop(['Price'],axis=1)
y_train=a3['Price']
```
